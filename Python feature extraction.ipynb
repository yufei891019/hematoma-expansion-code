```python
import os
import SimpleITK as sitk 
import radiomics#
from radiomics import featureextractor

import numpy as np
import pandas as pd
from tqdm import tqdm
import re
```

```python
# 1) 
import radiomics
from radiomics import featureextractor
para_path = r'E:\slicer05mm\original_data\GZ_ROI_05add/exampleCT.yaml' 
extractor = featureextractor.RadiomicsFeatureExtractor(para_path) 

# 2) 
dataset_dir = r'E:\slicer05mm\original_data\GZ_ROI_05add'#
samples_list = os.listdir(dataset_dir) # 

error_samples = [] 
dataset_result_df = pd.DataFrame([])

for sample in tqdm(samples_list):
    
    sample_dir = os.path.join(dataset_dir,sample)
    image_filepath = os.path.join(sample_dir,'CT.nrrd')
    seg_filepath = os.path.join(sample_dir,'Segmentation.seg.nrrd')
    
    try:
        image_sitk = sitk.ReadImage(image_filepath)
    except:
        error_samples  += [sample]
        print("Skip error sample( Error while reading 1.nrrd reading ): %s"%sample)
        continue
        
    try:
        seg_sitk = sitk.ReadImage(seg_filepath)
    except:
        error_samples  += [sample]
        print("Skip error sample( Error while reading Segmentation.seg.nrrd  ): %s"%sample)
        continue
    
    try:
        result = extractor.execute(image_sitk,seg_sitk) 
        
        pick_result = dict([])
        for key, value in result.items(): 
            if 'diagnostics' in key:
                continue
            else:
                pick_result[key] = value
        sample_result_df = pd.DataFrame([pick_result],index=[sample])
        dataset_result_df = pd.concat([dataset_result_df,sample_result_df],axis=0)
    except:
        error_samples  += [sample]
        print("Skip error sample( label error ): %s"%sample)
        continue
display(dataset_result_df)    
```

```python
save_dir = r"E:\Python05\FZ168\ER\four\返回ER\代码"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
filepath = os.path.join(save_dir,'GYadd_radiomics.csv')
dataset_result_df.to_csv( filepath )
```
